## Visualisierung der Dichtematrix

### Beispiel GHZ-Zustand

```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from qiskit.visualization import plot_state_city

import numpy as np
```

```python
qc_GHZ = QuantumCircuit(3)
qc_GHZ.h(0)
qc_GHZ.cx(0,1)
qc_GHZ.cx(0,2)
qc_GHZ.draw('mpl')
```

```python
rho_GHZ = qi.DensityMatrix.from_instruction(qc_GHZ)
rho_GHZ.draw('latex', prefix='\\rho_{GHZ} = ')
```

```python
plot_state_city(rho_GHZ.data, title='Density Matrix')
```

### Beispiel W-Zustand

```python
qc_W = QuantumCircuit(3)

qc_W.ry( 2*np.arccos( 1/np.sqrt(3) ), 0 )
qc_W.ch(0, 1)
qc_W.cx(1,2)
qc_W.cx(0,1)
qc_W.x(0)
qc_W.draw('mpl')
```

```python
rho_W = qi.DensityMatrix.from_instruction(qc_W)
rho_W.draw('latex', prefix='\\rho_{W} = ')
```

```python
plot_state_city(rho_W.data, title='Density Matrix')
```

### Beispiel Separaierbarer Zustand

```python
qc_separable = QuantumCircuit(3)

qc_separable.h(0)
qc_separable.x(1)
qc_separable.ry( 2*np.pi/3, 2)
qc_separable.draw('mpl')
```
